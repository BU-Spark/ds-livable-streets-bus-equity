Hi, I'm Cindy Zhang and the following code is my visualization for Cambridge. The data is from data.census.gov and it's not very organized so I transferred it into pandaframe format manually(using chatGPT).

The dataset I'm using is [S0802 Means of Transportation to Work by Selected Characteristics](https://data.census.gov/table/ACSST1Y2022.S0802?q=commute+in+cambridge)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data = {
    "Label (Grouping)": [ "Car, truck, or van", "Drove alone", "Carpooled", "In 2-person carpool", 
                         "In 3-person carpool", "In 4-or-more person carpool", "Public transportation (excluding taxicab)", 
                         "Walked", "Bicycle", "Taxicab, motorcycle, or other means", "Worked from home"],
    "Total Estimate": [24.6, 21.8, 2.8, 2.5, 0.3, 0.1, 19.1, 17.4, 7.2, 1.7, 30.1]
}

df = pd.DataFrame(data)

# Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(x="Total Estimate", y="Label (Grouping)", data=df, palette="viridis")
plt.title('Number of Workers and Means of Transportation in Cambridge city, Massachusetts')
plt.xlabel('Percentage(%)')
plt.ylabel('Means of Transportation')
plt.show()
```

```python

# Raw data to DataFrame
data = {
    "Label (Grouping)": ["16 to 19 years", "20 to 24 years", "25 to 44 years", "45 to 54 years", "55 to 59 years", 
                         "60 years and over", "Male", "Female", "White", "Black or African American", 
                         "Asian", "Two or more races", "Hispanic or Latino origin (of any race)", 
                         "White alone, not Hispanic or Latino"],
    "Drove Alone": [1.7, 13.2, 49.0, 20.2, 3.8, 12.0, 51.9, 48.1, 62.3, 20.1, 9.9, 5.9, 5.1, 61.5],
    "Carpooled": [7.6, 4.8, 37.4, 25.2, 3.2, 21.8, 45.5, 54.5, 54.3, 20.9, 13.6, 11.3, 12.3, 52.3],
    "Public Transportation": [1.0, 25.9, 55.1, 6.1, 3.2, 8.7, 35.0, 65.0, 55.3, 12.3, 16.6, 9.6, 11.5, 54.8]
}

df = pd.DataFrame(data)

# Data separation
age_data = df.iloc[:6, :]
sex_data = df.iloc[6:8, :]
race_data = df.iloc[8:, :]

# Plotting Age Data
plt.figure(figsize=(10, 6))
age_melted = age_data.melt(id_vars="Label (Grouping)", value_vars=["Drove Alone", "Carpooled", "Public Transportation"], 
                           var_name="Transportation", value_name="Percentage")
sns.barplot(x="Percentage", y="Label (Grouping)", hue="Transportation", data=age_melted, palette="viridis")
plt.title('Transportation Percentage by Age in Cambridge city, Massachusetts')
plt.xlabel('Percentage')
plt.ylabel('Age Group')
plt.legend(title="Means of Transportation")
plt.show()

# Plotting Sex Data
plt.figure(figsize=(8, 4))
sex_melted = sex_data.melt(id_vars="Label (Grouping)", value_vars=["Drove Alone", "Carpooled", "Public Transportation"], 
                           var_name="Transportation", value_name="Percentage")
sns.barplot(x="Percentage", y="Label (Grouping)", hue="Transportation", data=sex_melted, palette="viridis")
plt.title('Transportation Percentage by Sex in Cambridge city, Massachusetts')
plt.xlabel('Percentage')
plt.ylabel('Sex')
plt.legend(title="Means of Transportation")
plt.show()

# Plotting Race Data
plt.figure(figsize=(10, 8))
race_melted = race_data.melt(id_vars="Label (Grouping)", value_vars=["Drove Alone", "Carpooled", "Public Transportation"], 
                             var_name="Transportation", value_name="Percentage")
sns.barplot(x="Percentage", y="Label (Grouping)", hue="Transportation", data=race_melted, palette="viridis")
plt.title('Transportation Percentage by Race in Cambridge city, Massachusetts')
plt.xlabel('Percentage')
plt.ylabel('Race')
plt.legend(title="Means of Transportation")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 设置风格
sns.set_theme(style="whitegrid")

# Make sure the year column is sorted in ascending order
final_df["Year"] = pd.to_numeric(final_df["Year"])
final_df = final_df.sort_values("Year")

# 1. Gender vs. Public Transportation Usage
plt.figure(figsize=(10, 6))
sns.lineplot(data=final_df[final_df["Label (Grouping)"].isin(["Male", "Female"])], 
             x="Year", y="Public Transportation", hue="Label (Grouping)")
plt.title("Public Transportation Percentage by Sex in Cambridge MA")
plt.show()

# 2. Race vs. Public Transportation Usage
plt.figure(figsize=(10, 6))
sns.lineplot(data=final_df[final_df["Label (Grouping)"].isin(["White", "Black or African American", "Asian", "Hispanic or Latino origin (of any race)", "White alone, not Hispanic or Latino"])], 
             x="Year", y="Public Transportation", hue="Label (Grouping)")
plt.title("Public Transportation Percentage by Race in Cambridge MA")
plt.show()

# 3. Age vs. Public Transportation Usage (20-24 and 25-44 years)
plt.figure(figsize=(10, 6))
sns.lineplot(data=final_df[final_df["Label (Grouping)"].isin(["20 to 24 years", "25 to 44 years"])], 
             x="Year", y="Public Transportation", hue="Label (Grouping)")
plt.title("Public Transportation Percentage by Age in Cambridge MA")
plt.show()
```
