```python
# List of dataframes that have been read
dataframes = {
    # '2014': df_2014,
    # '2015': df_2015,
    '2016': df_2016,
    '2017': df_2017,
    '2018': df_2018,
}

# Initializes a dictionary to store averages
average_jwmnp_values = {}

# Iterate over each DataFrame
for year, df in dataframes.items():
    # Filter out the rows with PUMA values in common pumas
    # df_common_pumas = df[df['PUMA'].isin(common_pumas)]

    # Filter the data whose JWTRNS is 02
    df_filtered = df[df['JWTR'] == 2]

    # Group each PUMA value
    grouped = df_filtered.groupby('PUMA')

    # Traverse each PUMA group
    for puma, group in grouped:
        # Iterate over the value of RAC1P
        for rac1p_value in [1, 2]:
            # Apply further screening criteria
            condition = group['RAC1P'] == rac1p_value

            # Calculate the average
            average_jwmnp = group[condition]['JWMNP'].mean()

            # Store results
            average_jwmnp_values[f"{year}, PUMA={puma}, RAC1P={rac1p_value}"] = average_jwmnp
```

```python
import pandas as pd

# Suppose average_jwmnp_values is the dictionary you created earlier, with the keys containing year, PUMA, and RAC1P
# average_jwmnp_values = {...}

# Convert the dictionary format to fit the DataFrame
data_for_df = []
for key, value in average_jwmnp_values.items():
    year, puma_rac1p = key.split(", PUMA=")
    puma, rac1p = puma_rac1p.split(", RAC1P=")
    data_for_df.append({'Year': int(year), 'PUMA': int(puma), 'RAC1P': int(rac1p), 'Average JWMNP': value})

# Create a DataFrame
df_average_jwmnp = pd.DataFrame(data_for_df)

# Displays the DataFrame
df_average_jwmnp
```

```python
# Invert the dictionary so that it maps from values to keys
puma_dict_2010_reversed = {v: k for k, v in puma_dict_2010.items()}
puma_dict_2020_reversed = {v: k for k, v in puma_dict_2020.items()}

# Define replacement function
def replace_puma(row):
    year = int(row['Year'])
    puma = row['PUMA']

    if year <= 2019:
        return puma_dict_2010_reversed.get(puma, puma)  # Replace with the inverted puma_dict_2010
    else:
        return puma_dict_2020_reversed.get(puma, puma)  # Replace with the inverted puma_dict_2020

# Apply the replacement function to each line of df_average_jwmnp
df_average_jwmnp['PUMA'] = df_average_jwmnp.apply(replace_puma, axis=1)

# View the updated DataFrame
df_average_jwmnp
```

```python
# Gets the unique PUMA value for each year
unique_pumas_by_year = df_average_jwmnp.groupby('Year')['PUMA'].unique()

# Find the PUMA values for all years
common_pumas = set(unique_pumas_by_year.iloc[0])
for pumas in unique_pumas_by_year[1:]:
    common_pumas.intersection_update(pumas)

# Filter out the rows that have these common PUMA values
df_common_pumas = df_average_jwmnp[df_average_jwmnp['PUMA'].isin(common_pumas)]

# Displays the filtered DataFrame
df_common_pumas
```

```python
# Find the PUMA value for the Average JWMNP line containing NaN
pumas_with_nan = df_common_pumas[df_common_pumas['Average JWMNP'].isna()]['PUMA'].unique()

# Filter out all rows that do not contain these PUMA values
df_filtered = df_common_pumas[~df_common_pumas['PUMA'].isin(pumas_with_nan)]

# Displays the filtered DataFrame

df_filtered
```

```python
# Create a list to store the difference for each PUMA
data_for_df = []
puma_rac1p_avg = df_filtered.groupby(['PUMA', 'RAC1P'])['Average JWMNP'].mean().reset_index()
# Loop through each PUMA
for puma in puma_rac1p_avg['PUMA'].unique():
    # Filter data for a specific PUMA
    df_puma = puma_rac1p_avg[puma_rac1p_avg['PUMA'] == puma]

    # Make sure both RAC1P values are present
    if 1 in df_puma['RAC1P'].values and 2 in df_puma['RAC1P'].values:
        avg_jwmnp_rac1p1 = df_puma[df_puma['RAC1P'] == 1]['Average JWMNP'].iloc[0]
        avg_jwmnp_rac1p2 = df_puma[df_puma['RAC1P'] == 2]['Average JWMNP'].iloc[0]

        # Calculated difference
        diff = avg_jwmnp_rac1p2 - avg_jwmnp_rac1p1
        data_for_df.append({'PUMA': puma, 'Difference': diff})

# Create a DataFrame
df_puma_diff = pd.DataFrame(data_for_df)

# Displays the DataFrame
df_puma_diff
```

```python
def simplify_puma_name(name):
    name_parts = name.split('--')[0].split('&')
    simplified_names = [part.split('(')[0].strip() for part in name_parts]
    return ', '.join(simplified_names)
df=df_puma_diff
df['Simplified PUMA'] = df['PUMA'].apply(simplify_puma_name)

# Refactor the DataFrame to group regions with the same difference
grouped_data = {}
for _, row in df.iterrows():
    diff = row['Difference']
    puma = row['Simplified PUMA']
    if diff in grouped_data:
        grouped_data[diff].append(puma)
    else:
        grouped_data[diff] = [puma]

# Create a new DataFrame
new_df_data = {'Difference': [], 'Regions': []}
for diff, pumas in grouped_data.items():
    new_df_data['Difference'].append(diff)
    new_df_data['Regions'].append(', '.join(pumas))

new_df = pd.DataFrame(new_df_data)

# Output the new DataFrame
print(new_df)
```

```python
# Calculate the Average JWMNP for each PUMA and RAC1P combination
puma_rac1p_avg = df_filtered.groupby(['PUMA', 'RAC1P'])['Average JWMNP'].mean().reset_index()

# The dictionary is initialized to store the maximum and minimum differences for each PUMA
puma_diff = {}

# Walk through each PUMA
for puma in puma_rac1p_avg['PUMA'].unique():
    # Filter data for a specific PUMA
    df_puma = puma_rac1p_avg[puma_rac1p_avg['PUMA'] == puma]

    # Make sure there are two RAC1P values
    if df_puma.shape[0] == 2:
        # Calculated difference
        diff = abs(df_puma.iloc[0]['Average JWMNP'] - df_puma.iloc[1]['Average JWMNP'])
        puma_diff[puma] = diff

# Calculate the maximum and minimum differences
max_diff_puma = max(puma_diff, key=puma_diff.get)
min_diff_puma = min(puma_diff, key=puma_diff.get)

max_diff = puma_diff[max_diff_puma]
min_diff = puma_diff[min_diff_puma]

print(f"PUMA have maximum difference: {max_diff_puma},\n Difference: {(max_diff * 2 * 5 * 50) / 60}")
print(f"PUMA have minimum difference: {min_diff_puma},\n Difference: {(min_diff * 2 * 5 * 50) / 60}")
```

```python
import pandas as pd

# Load the CSV file
file_path = './df_puma_diff.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
zip_codes = []
differences = []

for _, row in df.iterrows():
    # Splitting ZIP codes by space and comma, then stripping any whitespace
    codes = [code.strip() for code in str(row['PUMA']).replace(',', ' ').split()]
    zip_codes.extend(codes)
    differences.extend([row['Difference']] * len(codes))

# Creating a new DataFrame
df_separated = pd.DataFrame({
    'ZIP Code': zip_codes,
    'Difference': differences
})
df_separated["Difference"]=(df_separated["Difference"]* 2 * 5 * 50) / 60
df_separated.to_csv("zip.csv")
```
