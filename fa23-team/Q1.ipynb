## Roxbury

```python
black_2020 = pd.read_csv('Roxbury_data_2020b.csv')
black_2015 = pd.read_csv('Roxbury_data_2015b.csv')
black_2010 = pd.read_csv('Roxbury_data_2010b.csv')
black_2021 = pd.read_csv('Roxbury_data_2021b.csv')
white_2020 = pd.read_csv('Roxbury_data_2020w.csv')
white_2015 = pd.read_csv('Roxbury_data_2015w.csv')
white_2010 = pd.read_csv('Roxbury_data_2010w.csv')
white_2021 = pd.read_csv('Roxbury_data_2021w.csv')
```

```python
# List of dataframes for convenience
data = [black_2020, black_2015, black_2010, black_2021, white_2020, white_2015, white_2010, white_2021]

# drop the 'Margin of Error' columns
for df in data:
    margin_error_cols = [col for col in df.columns if 'Margin of Error' in col]
    df.drop(columns=margin_error_cols, inplace=True)
```

```python
# Convert all columns of the third row to numeric, ignoring non-numeric values
sum_black_2020 = pd.to_numeric(black_2020.iloc[3], errors='coerce').sum()-3
sum_black_2015 = pd.to_numeric(black_2015.iloc[3], errors='coerce').sum()-3
sum_black_2010 = pd.to_numeric(black_2010.iloc[3], errors='coerce').sum()-3
sum_black_2021 = pd.to_numeric(black_2021.iloc[3], errors='coerce').sum()-3
sum_white_2020 = pd.to_numeric(white_2020.iloc[3], errors='coerce').sum()-3
sum_white_2015 = pd.to_numeric(white_2015.iloc[3], errors='coerce').sum()-3
sum_white_2010 = pd.to_numeric(white_2010.iloc[3], errors='coerce').sum()-3
sum_white_2021 = pd.to_numeric(white_2021.iloc[3], errors='coerce').sum()-3
```

```python
print("Sum of 5th row for Black 2020 Data:", sum_black_2020)
print("Sum of 5th row for Black 2015 Data:", sum_black_2015)
print("Sum of 5th row for Black 2010 Data:", sum_black_2010)
print("Sum of 5th row for Black 2021 Data:", sum_black_2021)
print("Sum of 5th row for White 2020 Data:", sum_white_2020)
print("Sum of 5th row for White 2015 Data:", sum_white_2015)
print("Sum of 5th row for White 2010 Data:", sum_white_2010)
print("Sum of 5th row for White 2021 Data:", sum_white_2021)
```

```python
def percentage_change(old, new):
    if old == 0:
        return 0  # Avoid division by zero
    return ((new - old) / old) * 100

# Black Data
black_changes = {
    "2010-2015": percentage_change(sum_black_2010, sum_black_2015),
    "2015-2020": percentage_change(sum_black_2015, sum_black_2020),
    "2020-2021": percentage_change(sum_black_2020, sum_black_2021)
}

# White Data
white_changes = {
    "2010-2015": percentage_change(sum_white_2010, sum_white_2015),
    "2015-2020": percentage_change(sum_white_2015, sum_white_2020),
    "2020-2021": percentage_change(sum_white_2020, sum_white_2021)
}

# Print the results
print("Black Data Percentage Changes:")
for period, change in black_changes.items():
    print(f"{period}: {change:.2f}%")

print("\nWhite Data Percentage Changes:")
for period, change in white_changes.items():
    print(f"{period}: {change:.2f}%")
```

```python
# Calculate total for each year
total_2020 = sum_black_2020 + sum_white_2020
total_2015 = sum_black_2015 + sum_white_2015
total_2010 = sum_black_2010 + sum_white_2010
total_2021 = sum_black_2021 + sum_white_2021

# Calculate percentage for Black
black_percentages = [
    (sum_black_2010 / total_2010) * 100,
    (sum_black_2015 / total_2015) * 100,
    (sum_black_2020 / total_2020) * 100,
    (sum_black_2021 / total_2021) * 100
]

# Calculate percentage for White
white_percentages = [
    (sum_white_2010 / total_2010) * 100,
    (sum_white_2015 / total_2015) * 100,
    (sum_white_2020 / total_2020) * 100,
    (sum_white_2021 / total_2021) * 100
]

# Years
years = ['2010', '2015', '2020', '2021']

# Plotting for Black
plt.figure(figsize=(10, 6))
plt.plot(years, black_percentages, marker='o', color='black')
plt.title('Black Percentage of Sums Over the Years')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

# Plotting for White
plt.figure(figsize=(10, 6))
plt.plot(years, white_percentages, marker='o', color='grey')
plt.title('White Percentage of Sums Over the Years')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
data = {
    'Year': ['2010', '2015', '2020', '2021'],
    'Black': [2477, 3438, 3446, 3052],
    'White': [1755, 2448, 2900, 2592]
}

df = pd.DataFrame(data)
df.set_index('Year', inplace=True)
bar_width = 0.35

index = np.arange(len(df))

bars1 = plt.bar(index, df['Black'], bar_width, label='Black', color='black', alpha=0.7)
bars2 = plt.bar(index + bar_width, df['White'], bar_width, label='White', color='grey', alpha=0.7)

plt.plot(index, df['Black'], marker='o', color='red', label='Black (Line)')
plt.plot(index + bar_width, df['White'], marker='o', color='blue', label='White (Line)')

def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.annotate('{}'.format(height),
                     xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3),  
                     textcoords="offset points",
                     ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)
plt.legend()
plt.title('Public Transportation Usage: Black vs White Riders (2010-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Riders')
plt.xticks(index + bar_width / 2, df.index)
plt.tight_layout()
plt.show()
```

```python
print("Sum of 5th row for Black 2020 Data:", sum_black_2020)
print("Sum of 5th row for Black 2015 Data:", sum_black_2015)
print("Sum of 5th row for Black 2010 Data:", sum_black_2010)
print("Sum of 5th row for Black 2021 Data:", sum_black_2021)
print("Sum of 5th row for White 2020 Data:", sum_white_2020)
print("Sum of 5th row for White 2015 Data:", sum_white_2015)
print("Sum of 5th row for White 2010 Data:", sum_white_2010)
print("Sum of 5th row for White 2021 Data:", sum_white_2021)
```

```python
def percentage_change(old, new):
    if old == 0:
        return 0  
    return ((new - old) / old) * 100

# Black Data
black_changes = {
    "2010-2015": percentage_change(sum_black_2010, sum_black_2015),
    "2015-2020": percentage_change(sum_black_2015, sum_black_2020),
    "2020-2021": percentage_change(sum_black_2020, sum_black_2021)
}

# White Data
white_changes = {
    "2010-2015": percentage_change(sum_white_2010, sum_white_2015),
    "2015-2020": percentage_change(sum_white_2015, sum_white_2020),
    "2020-2021": percentage_change(sum_white_2020, sum_white_2021)
}

# Print the results
print("Black Data Percentage Changes:")
for period, change in black_changes.items():
    print(f"{period}: {change:.2f}%")

print("\nWhite Data Percentage Changes:")
for period, change in white_changes.items():
    print(f"{period}: {change:.2f}%")
```

```python
# Calculate total for each year
total_2020 = sum_black_2020 + sum_white_2020
total_2015 = sum_black_2015 + sum_white_2015
total_2010 = sum_black_2010 + sum_white_2010
total_2021 = sum_black_2021 + sum_white_2021

# Calculate percentage for Black
black_percentages = [
    (sum_black_2010 / total_2010) * 100,
    (sum_black_2015 / total_2015) * 100,
    (sum_black_2020 / total_2020) * 100,
    (sum_black_2021 / total_2021) * 100
]

# Calculate percentage for White
white_percentages = [
    (sum_white_2010 / total_2010) * 100,
    (sum_white_2015 / total_2015) * 100,
    (sum_white_2020 / total_2020) * 100,
    (sum_white_2021 / total_2021) * 100
]

# Plotting
years = ['2010', '2015', '2020', '2021']
plt.figure(figsize=(10, 6))
plt.plot(years, black_percentages, marker='o', label='Black')
plt.plot(years, white_percentages, marker='o', label='White')

plt.title('Percentage of Sums Over the Years')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
data = {
    'Year': ['2010', '2015', '2020', '2021'],
    'Black': [7855, 8594, 8274, 6318],
    'White': [4428, 5425, 6278, 5576]
}
df = pd.DataFrame(data)
df.set_index('Year', inplace=True)

bar_width = 0.35

index = np.arange(len(df))

bars1 = plt.bar(index, df['Black'], bar_width, label='Black', color='black', alpha=0.7)
bars2 = plt.bar(index + bar_width, df['White'], bar_width, label='White', color='grey', alpha=0.7)

plt.plot(index, df['Black'], marker='o', color='red', label='Black (Line)')
plt.plot(index + bar_width, df['White'], marker='o', color='blue', label='White (Line)')

def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.annotate('{}'.format(height),
                     xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3), 
                     textcoords="offset points",
                     ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)
plt.legend()
plt.title('Public Transportation Usage: Black vs White Riders (2010-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Riders')
plt.xticks(index + bar_width / 2, df.index)
plt.tight_layout()
plt.show()
```
