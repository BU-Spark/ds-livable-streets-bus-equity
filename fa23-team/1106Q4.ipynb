```python
import csv
import glob
import os

def main():
    #folder_path = 'path/to/your/folder'
    output_file_path = 'output.csv'
    
    # Header for the output CSV file
    output_header = ['File', 'Sum']
    output_rows = []

    neighborhoods = ['Mattapan', 'Roxbury', 'Dorcherter']
    colors = ['black', 'white']
    years = range(2016, 2022)

    for neighborhood in neighborhoods:
        for color in colors:
            for year in years:
                file_pattern = f'{neighborhood}_{color}_{year}.csv'
                for file_path in glob.glob(file_pattern):
                    with open(file_path, newline='', encoding='utf-8') as csvfile:
                        csv_reader = csv.reader(csvfile)
                        rows = list(csv_reader)
                        fourth_row = rows[3]  # Get the fourth row
                        
                        # Get the indices of columns ending with 'Estimate'
                        estimate_columns_indices = [i for i, col in enumerate(rows[0]) if col.endswith('Estimate')]
                        
                        # Sum the values in the fourth row for the selected columns
                        total_estimate = sum(int(fourth_row[i]) for i in estimate_columns_indices)
                        
                        # Prepare the output row
                        output_rows.append([file_path, total_estimate])
    
    # Write the sums to the output CSV file
    with open(output_file_path, mode='w', newline='', encoding='utf-8') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(output_header)
        csv_writer.writerows(output_rows)

if __name__ == "__main__":
    main()
```

```python
import csv
import glob
import os

def main():
    # folder_path = 'path/to/your/folder'
    output_file_path = 'output_time.csv'
    
    # Header for the output CSV file
    output_header = ['File', 'Row Index', 'Sum']  # Added 'Row Index'-the header
    output_rows = []

    neighborhoods = ['Mattapan', 'Roxbury', 'Dorchester']
    years = range(2019, 2022)

    for neighborhood in neighborhoods:
            for year in years:
                file_pattern = f'{neighborhood}_time_{year}.csv'
                for file_path in glob.glob(file_pattern):
                    with open(file_path, newline='', encoding='utf-8') as csvfile:
                        csv_reader = csv.reader(csvfile)
                        rows = list(csv_reader)
                        
                        # Get the indices of columns ending with 'Estimate'
                        estimate_columns_indices = [i for i, col in enumerate(rows[0]) if col.endswith('Estimate')]
                        
                        for row_index, row in enumerate(rows[1:], start=1):  # Start enumeration from 1-skip the header row
                            # Sum the values in the current row for the selected columns
                            total_estimate = sum(int(row[i]) for i in estimate_columns_indices)/len(estimate_columns_indices)
                            
                            # Prepare the output row
                            output_rows.append([file_path, row_index, total_estimate])  # Added row_index-the output row
    
    # Write the sums to the output CSV file
    with open(output_file_path, mode='w', newline='', encoding='utf-8') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(output_header)
        csv_writer.writerows(output_rows)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('output.csv')

# Split the data
data_before_2020 = data[data['year'] < 2020]
data_after_2020 = data[data['year'] >= 2020]

def calculate_average_commute_times(data):
    # Group by city and color, then calculate the mean sum for each group
    return data.groupby(['city', 'color'])['population'].mean().unstack()

# Calculate average commute times
average_before_2020 = calculate_average_commute_times(data_before_2020)
average_after_2020 = calculate_average_commute_times(data_after_2020)

# Compute percentage change
percentage_change = ((average_after_2020 - average_before_2020) / average_before_2020) * 100

# Visualize original data
g = sns.catplot(data=data, x='year', y='population', hue='color', col='city', kind='bar', errorbar=None)
g.fig.suptitle('Bus Commute Times by Year, City, and Color', y=1.02)  # Adjust title position
plt.show()

# Visualize percentage change
percentage_change.plot(kind='barh', stacked=True)
plt.title('Percentage Change in Average Commute Times Due to Covid-19')
#plt.ylabel('Percentage Change')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data.csv' is your dataset file
df = pd.read_csv('output_time2.csv')

# Mapping from time to descriptive labels
time_labels = {
    2: '< 10',
    3: '10-14',
    4: '15-19',
    5: '20-24',
    6: '25-29',
    7: '30-34',
    8: '35-44',
    9: '45-59',
    10: '> 60'
}

# Now plot the data for each city and year
sns.set(style="whitegrid")  # For a nicer grid layout

for city in df['city'].unique():
    city_data = df[df['city'] == city]
    plt.figure(figsize=(14, 7))
    
    # Create a bar plot for the city across all years
    sns.barplot(data=city_data, x='time', y='population', hue='year', palette='viridis')
    
    # Set the x-labels to the time categories using the labels dictionary
    plt.xticks(range(len(time_labels)), list(time_labels.values()), rotation=45)
    
    plt.title(f'Average Population Commute Times in {city} (2019-2021)')
    plt.xlabel('Commute Time Categories')
    plt.ylabel('Average Population')
    plt.legend(title='Year')
    plt.tight_layout()
    plt.show()
```
